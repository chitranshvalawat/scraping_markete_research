```python
df["Individual Size (gm)"] = df["Individual Size (gm)"].str.replace(r'[A-Z]|[a-z]', '')
```

```python
df["Price (SAR)"] = df["Price (SAR)"].str.replace(r'[A-Z]|[a-z]', '')
```

```python
df["Price (SAR)"] = df["Price (SAR)"].str.replace('/', '')
```

```python
df["Individual Size (gm)"] = df["Individual Size (gm)"].str.replace('/', '')
```

```python
df["Individual Size (gm)"] = df["Individual Size (gm)"].str.replace('/', '')
```

```python
for i in range(len(df)) :
    print("index:" + str(i))
    print(df.loc[i, "Individual Size (gm)"])
    if "." in df.loc[i, "Individual Size (gm)"]:
        print(df.loc[i, "Individual Size (gm)"])
        df.loc[i, "Individual Size (gm)"] = df.loc[i, "Individual Size (gm)"].split(".")[0]
        print(df.loc[i, "Individual Size (gm)"])
```

```python
df['Price (SAR)'] = pd.to_numeric(df['Price (SAR)'],downcast="float")
```

```python
df['Individual Size (gm)'] = pd.to_numeric(df['Individual Size (gm)'],downcast="float")
```

```python
for i in range(len(df)) :
    df.loc[i,["SAR/g"]] = (df.loc[i,"Price (SAR)"]).astype(int) / (df.loc[i,"Individual Size (gm)"]).astype(int)
    if df.loc[i, "IS_Combo"] == "yes" or df.loc[i, "IS_Combo"] == "Yes" or df.loc[i, "IS_Combo"] == "YES":
        df.loc[i,["SAR/g"]] = df.loc[i,["SAR/g"]] /df.loc[i,["If Combo - Size"]] 
        print(df.loc[i,"Individual Size (gm)"])
    
```
